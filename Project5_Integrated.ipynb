# Testing the Hypotheses

Description: The following hypothesis will be examined: 
a.Average user ratings of the Xbox One and PC platforms are the same.
b.Average user ratings for the Action and Sports genres are different.

In order to examine these hypothesis it will be necessary to complete a two sample indendent t-test for both. It is common practice to take a signfiance level of 0.05, therefore that will be the signficance level used for this analysis. The aim will be to determine if the null hypothesis can be rejected or cannot be rejected. In statistics, the null hypothesis is the assumption that there is no difference between the groups being tested. Whereas the alternative hypothesis is the assumption that there is a a difference between the groups being tested.

```python
#Average user ratings of the Xbox One and PC platforms are the same

#Filter data 
xbox_one_scores=df[df['platform']=='XOne']['user_score'].dropna().astype(float)
pc_scores=df[df['platform']=='PC']['user_score'].dropna().astype(float)

#Perform two sample t-test 
t_stat, p_value = st.ttest_ind(xbox_one_scores,pc_scores,equal_var=True)

#Interpretation 
alpha=0.05
print(f"T-statistic:{t_stat}")
print(f"P-value:{p_value}")
      
if p_value < alpha:
      print("Reject null hypothesis: Average user ratings of the Xbox One and PC platforms are not the same. ")
else:
      print("Fail to reject the null hypothesis: Average user ratings of the Xbox One and PC platforms are the same.")
```

```python
#Average user ratings for the Action and Sports genres are different 
#Filter data for action and sports genres 
action_scores=df[df['genre']=='Action']['user_score'].dropna().astype(float)
sports_scores=df[df['genre']=='Sports']['user_score'].dropna().astype(float)

#Perform two sample t-test 
t_stat, p_value = st.ttest_ind(action_scores,sports_scores,equal_var=True)

#Interpretation 
alpha=0.05
print(f"T-statistic:{t_stat}")
print(f"P-value:{p_value}")
      
if p_value < alpha:
      print("Reject null hypothesis: Average user ratings for the Action and Sports genres are different. ")
else:
      print("Fail to reject the null hypothesis: Average user ratings for the Action and Sports genres are the same.")
```

Description: The statistical analysis illustrates that the null hypothesis for "Average user rating for the Action and Sports genres" cannot be rejected. Moreover, the null hypothesis for "Average user ratings of the Xbox One and PC platforms are the same" cannot be rejected. This is corroborated by earlier findings that user rating is not signficant in impacting consumer purchasing. This outcome proves that marketing campaigns should not target user reviews for their advertising campaigns as it will not make a signficant difference.
